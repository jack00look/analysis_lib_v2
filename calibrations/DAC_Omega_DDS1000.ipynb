```python
import numpy as np
import matplotlib.pyplot as plt
#imprt pchi interpolat
from scipy.interpolate import PchipInterpolator
```

```python
# load from txt
data = np.loadtxt('/home/rick/labscript-suite/userlib/analysislib/analysislib_v2/calibrations/DAC_Omega_DDS1000_data.txt')
V = data[:,0]
f = data[:,1]
transfer_func = PchipInterpolator(f,V,extrapolate=False)
```

```python
plt.plot(f,V,'o-')
plt.xlim(0,1000)
plt.ylim(0,0.6)

interpolate_test_f = np.linspace(170,1000,100)
interpolate_test_V = transfer_func(interpolate_test_f)
plt.plot(interpolate_test_f,interpolate_test_V,'x-')
plt.xlabel('DAC Voltage (V)')
plt.ylabel('Rabi Frequency (kHz)')
plt.title('DAC Voltage to Rabi Frequency Calibration')
plt.grid()
plt.show()
```
