```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10  # number of samples
x = np.logspace(np.log10(0.01), np.log10(3), N)
print(x)
plt.plot(np.linspace(0, 1, N), x)
```

```python
def autocorr(x):
    """
    Computes the autocorrelation of x.
    """
    result = np.correlate(x, x, mode='full')
    return result[result.size//2:] / result[result.size//2]
```

```python
period = 1.
x_axis = np.linspace(0.,3.,100)

sin_y = np.sin(2*np.pi*x/period)

autocorr_sin_y = autocorr(sin_y)

plt.plot(x_axis,autocorr_sin_y)
```
